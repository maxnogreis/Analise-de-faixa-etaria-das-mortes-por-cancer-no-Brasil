Taxas de mortalidade por câncer, brutas e ajustadas por idade pelas populações mundial e brasileira, por 100.000, segundo sexo, faixa etaria, localidade e por período selecionado. Dados Obtidos através do link https://mortalidade.inca.gov.br/MortalidadeWeb/pages/Modelo03/consultar.xhtml;jsessionid=108DAA6D9B822D6B6F079D7AB6FDC846#panelResultado

```python
import pandas as pd
import seaborn as sb
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90


filename = 'f1.csv'

colnames = ['Faixa Etária','Número de óbito de homens','Taxa específica para homens', 'Número de óbito de mulheres','Taxa específica para mulheres','Número total de óbitos', 'Taxa específica para todos']

df = pd.read_csv('f1.csv', sep=';' , encoding='latin-1' , names = colnames , skiprows=[0,1,2], skipfooter = 6)

df.head(12).plot(kind = "bar", figsize = (20,10), grid = (True), x=("Faixa Etária"), y =('Número de óbito de homens'),  subplots = True)
```

Podemos observar pelo gráfico acima, que a faixa etária entre homens com o maior índice de mortalidade por câncer no periodo de 1979 até 2015 foi a de 60 a 69 anos, com aproximadamente 600 mil mortes no período.

```python
df.head(12).plot(kind = "bar", figsize = (20,10), grid = (True), x=("Faixa Etária"), y =('Número de óbito de mulheres'),  subplots = True)
```

Podemos observar pelo gráfico acima, que a faixa etária entre mulheres com o maior índice de mortalidade por câncer no periodo de 1979 até 2015 também foi a de 60 a 69 anos, com aproximadamente 450 mil mortes no período.

```python
df.head(12).plot(kind = "bar", figsize = (20,10), grid = (True), x=("Faixa Etária"), y =('Número total de óbitos'),  subplots = True)
```

E, através deste gráfico com o total tanto de homens quanto de mulheres podemos constatar que a Faixa etária com maior risco é a de 60 a 69 anos com mais de 1 milhão de mortos no período.
